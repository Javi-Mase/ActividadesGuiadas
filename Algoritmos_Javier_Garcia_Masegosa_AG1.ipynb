## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

<font color="#27ae60">
El algoritmo voraz para el cambio de monedas funciona de manera perfecta y devuelve el óptimo global con nuestro sistema monetario del euro. Esto ocurre porque el Euro es un sistema canónico. En un sistema de este tipo, la regla heurística de elegir siempre la moneda de mayor valor posible funciona siempre.
</font>

<font color="#27ae60">
Sin embargo, la principal limitación teórica de las técnicas voraces es que no garantizan el óptimo global ante cualquier conjunto de restricciones. Si el sistema de monedas es dejara por asi decirlo de ser canónico, el algoritmo tomará una decisión localmente óptima.
</font>

<font color="#27ae60">
A continuación, demostramos esto con un sistema ficticio de monedas de valor <code>[11, 5, 1]</code> para dar un cambio de <code>15</code>. Veremos que el algoritmo voraz usa 5 monedas, mientras que la solución óptima real sería usar solo 3 (tres monedas de 5). Esto justifica teóricamente por qué para sistemas arbitrarios debemos abandonar los algoritmos voraces y utilizar Programación Dinámica.
</font>

```python
# Demostración del fallo del Algoritmo Voraz ante un sistema no canónico
def cambio_monedas_voraz_falla(cantidad, sistema_monedas):
    solucion = []
    valor_restante = cantidad
    
    # El algoritmo voraz siempre coge la moneda más grande primero
    for moneda in sistema_monedas:
        while valor_restante >= moneda:
            solucion.append(moneda)
            valor_restante -= moneda
            
    return solucion

# Definimos el escenario con un sistema de monedas diferente
monedas_ficticias = [11, 5, 1]
cantidad_cambio = 15

resultado_voraz = cambio_monedas_voraz_falla(cantidad_cambio, monedas_ficticias)

# Imprimimos la comparativa
print(f"Sistema de monedas disponible: {monedas_ficticias}")
print(f"Cantidad a devolver: {cantidad_cambio}\n")
print(f"Solución Algoritmo Voraz: {resultado_voraz}")
print(f"Total de monedas usadas con el algoritmo voraz: {len(resultado_voraz)}\n")
print(f"Solución óptima: [5, 5, 5]")
print(f"Total de monedas usadas con el algoritmo óptimo: 3")
```

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

# Actividades extra y mejoras

<font color="#27ae60">
A continuación, se implementa la solución al problema de encontrar los dos puntos más cercanos en un espacio 1D. Primero se presenta una solución por fuerza bruta con complejidad $O(N^2)$ y posteriormente una versión mejorada basándonos en la ordenación previa, con complejidad $O(N \log N)$.
</font>

```python


# Generamos una lista de 50 números aleatorios
puntos_1d = [random.randrange(1, 10000) for x in range(50)]


def puntos_cercanos_fuerza_bruta(lista_puntos):
    # Inicializamos con la máxima diferencia posible
    distancia_minima = float('inf')
    punto1 = None
    punto2 = None
    
    # Comparamos todos contra todos
    for i in range(len(lista_puntos)):
        for j in range(i + 1, len(lista_puntos)):
            distancia_actual = abs(lista_puntos[i] - lista_puntos[j])
            
            if distancia_actual < distancia_minima:
                distancia_minima = distancia_actual
                punto1 = lista_puntos[i]
                punto2 = lista_puntos[j]
                
    return punto1, punto2, distancia_minima

p1, p2, dist = puntos_cercanos_fuerza_bruta(puntos_1d)
print("Prueba de fuerza bruta:")
print(f"Los puntos más cercanos son {p1} y {p2} con una distancia de {dist}")
```

<font color="#27ae60">
Ahora procedemos a ordenar la lista para mejorar drásticamente el tiempo de ejecución. La ventaja de tener los números ordenados de menor a mayor es que los dos puntos más cercanos siempre van a estar juntos (serán adyacentes). De esta forma, en lugar de comparar todos los puntos contra todos como en la fuerza bruta (que nos daba una complejidad de $O(N^2)$), primero ordenamos la lista en $O(N \log N)$ y luego solo necesitamos hacer un recorrido lineal rápido comparando cada punto con su vecino de la derecha. Así logramos bajar la complejidad total del algoritmo a <b>$O(N \log N)$</b>.
</font>

```python
def puntos_cercanos_nlogn(lista_puntos):
    # Ordenar la lista
    lista_ordenada = sorted(lista_puntos)
    
    distancia_minima = float('inf')
    punto1 = None
    punto2 = None
    
    # Recorrer la lista comparando solo elementos adyacentes O(N)
    for i in range(len(lista_ordenada) - 1):
        distancia_actual = abs(lista_ordenada[i] - lista_ordenada[i+1])
        
        if distancia_actual < distancia_minima:
            distancia_minima = distancia_actual
            punto1 = lista_ordenada[i]
            punto2 = lista_ordenada[i+1]
            
    return punto1, punto2, distancia_minima

p1_mejor, p2_mejor, dist_mejor = puntos_cercanos_nlogn(puntos_1d)

print("Mejora a la fuerza bruta:")
print(f"Los puntos más cercanos son {p1_mejor} y {p2_mejor} con una distancia de {dist_mejor}")
```
